```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
#downloading the trained classifier XML file (haarcascade_frontalface_default.xml), which is available in OpenCvâ€™s GitHub repository and saveing it to our working location.
face_cascade=cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
```

```python
#downloading any images with faces
img = cv2.imread('/content/Indian_cricket_players_face.jpeg')
```

```python
#The detection works only on grayscale images. So it is important to convert the color image to grayscale.
gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)    
```

```python
faces = face_cascade.detectMultiScale(gray_img,scaleFactor=1.2,minNeighbors=4)

# Since some faces may be closer to the camera, they would appear bigger than the faces in the back. The scale factor compensates for this.
#The detection algorithm uses a moving window to detect objects. minNeighbors defines how many objects are detected near the current one before it declares the face found.

print("Total Faces:",len(faces))
print("Face Coordinates:",faces)
```

```python
#Draw Rectangle on Face

for x,y,w,h in faces:
    img=cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),3)
```

```python
cv2_imshow(img)
cv2.waitKey(0)
# cv2.destroyAllWindows()
```
