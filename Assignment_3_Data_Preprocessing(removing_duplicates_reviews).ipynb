```python
import pandas as pd
import re
import nltk
#nltk.download('all')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
nltk.download('averaged_perceptron_tagger')
```

```python
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
```

#1)	Read the reviews in review file and remove all duplicate reviews.

```python
review_df = pd.read_csv('/content/review_text.csv',names = ['messages'],header=0)
review_df
```

```python
len(review_df['messages'][1])
```

```python
cleaned_corpus = []
```

```python
ps = PorterStemmer()
```

```python
lm = WordNetLemmatizer()
```

```python
for i in range(0,len(review_df)):
  review = re.sub('[^A-Za-z]',' ',review_df['messages'][i])
  review = review.lower()
  review = review.split()
  review = [lm.lemmatize(word) for word in review if word not in stopwords.words('english')]
  review = ' '.join(review)
  cleaned_corpus.append(review)
```

```python
cleaned_corpus
```

```python
len(cleaned_corpus)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
X = cv.fit_transform(cleaned_corpus).toarray()
```

```python
X_df =pd.DataFrame(X)
X_df
```
