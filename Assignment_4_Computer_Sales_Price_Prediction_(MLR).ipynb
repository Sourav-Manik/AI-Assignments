#Data Preparation

```python
del computer_data['Unnamed: 0']
```

```python
computer_data.head(2)
```

```python
computer_data=pd.get_dummies(computer_data,columns=['cd','multi','premium'])
```

```python
computer_data.head(2)
```

#Model Building

Model building using Statsmodel

```python
X = computer_data.drop(labels= ['price'],axis = 1)           #Independent variables
y = computer_data[['price']]                                 #Dependent variable
```

```python
X.head()
```

#Data tranformation

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
scaled_X = std_scaler.fit_transform(X)
scaled_X = pd.DataFrame(data = scaled_X,columns = X.columns)
scaled_X
```

```python
scaled_X.mean()
```

```python
scaled_X.std()
```

#Model Training

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression() #Model Initialization
linear_model.fit(scaled_X,y)
```

```python
linear_model.intercept_, linear_model.coef_
```

#Model Testing

```python
y_pred = linear_model.predict(scaled_X)
y_pred
```

#Model Evaluation

```python
error = y-y_pred
error
```

Model Evaluation

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
```

```python
MSE = mean_squared_error(y,y_pred)
MSE
```

```python
#test data
RMSE = np.sqrt(mean_squared_error(y,y_pred))
RMSE
```

```python
R2_score = r2_score(y,y_pred)
R2_score
```
